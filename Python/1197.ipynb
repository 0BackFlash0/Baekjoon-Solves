```python
import sys
input = sys.stdin.readline

class Node:
    def __init__(self):

        self.parent = None

    def setParent(self, parent):
        self.parent = parent
    
    def getRoot(self, compress=True):
        if(self.parent==None):
            return self
        
        result = self.parent.getRoot()
        if(compress):
            self.setParent(result)

        return result

    def is_union(self, target_node):
        return self.getRoot() == target_node.getRoot()

    def union(self, target_node):
        if(not self.is_union(target_node)):
            target_node.getRoot().setParent(self.getRoot())

def min_spanning_graph(node_list, v, graph_list, e):
    graph_list.sort()

    graph_num = 0
    weight_sum = 0

    for graph in graph_list:
        weight = graph[0]
        a = node_list[graph[1]]
        b = node_list[graph[2]]

        if(a.is_union(b)):
            continue
        
        a.union(b)
        
        graph_num += 1
        weight_sum += weight

        if(graph_num==v-1):
            break

    return weight_sum



v, e = map(int, input().split())
node_list = [Node() for _ in range(v)]
graph_list = []
for _ in range(e):
    a, b, weight = map(int, input().split())

    graph_list.append((weight, a-1, b-1))

# print(graph_list)
print(min_spanning_graph(node_list, v, graph_list, e))
```
