```python
import sys
sys.setrecursionlimit(100000)

def div2or3(num):
    n = num
    i = 0
    while(num%2==0):
        num /= 2
        i += 1
    while(num%3==0):
        num /= 3
        i += 1

    if(num==1):
        return i
    else:
        return 0


def min_calc(num):

    memorize = [None]*(num+1)

    def store(num, result):
        if(memorize[num]==None):
            memorize[num] = result


    def calc(num):
        shortcut = div2or3(num)

        result = 0
        if(num==1):
            result =  0
        elif(memorize[num]!=None):
            result = memorize[num]
        elif(shortcut!=0):
            result =  shortcut
        else:
            minus_1 = calc(num-1)
            div_2 = calc(num//2) if num%2==0 else float("inf")
            div_3 = calc(num//3) if num%3==0 else float("inf")
            result = min([minus_1, div_2, div_3])+1

        store(num, result)
        return result
    
    return calc(num)


num = int(input())
print(min_calc(num))
# for i in range(50000, 99999):
    # print(min_calc(i))
```

```python
def min_calc(num):
    if(num==1):
        return 0

    memorize = []

    current = [num]
    next = []
    count = 0
    while True:

        while(current):
            n = current.pop()

            if(n not in memorize):
                memorize.append(n)

                next.append(n-1)
                if(n%3==0):
                    next.append(n//3)
                if(n%2==0):
                    next.append(n//2)

        current = next.copy()
        next = []
        count += 1
        if(1 in current):
            break
    
    return count


num = int(input())
print(min_calc(num))
# for i in range(50000, 99999):
    # print(min_calc(i))
```
