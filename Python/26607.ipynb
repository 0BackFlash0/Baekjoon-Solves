```python
import sys
input = sys.stdin.readline

def calc_stat(students, team_num, stat_sum):

    # Tree 탐색, DFS
    # 각 층은 한명의 학생을 의미
    # 학생들 : sort (내림차순)되어있음

    students.sort(reverse=True)

    pivot = stat_sum / 2
    min_stat_sum = team_num * stat_sum
    # 초기값
    tree = [[0, team_num, 0]]
    # Node 내부 값 : 지금까지 모집한 학생의 수, 파워의 합, 층(len)
    
    max_step = len(students)
    
    while len(tree)>0:
        node = tree.pop()

        stat_sum = node[0]
        rest_team = node[1]
        step = node[2]

        # 학생을 포함하지 않고 넘어감
        if(rest_team + step < max_step and stat_sum + sum(students[step+1:step+1+rest_team]) > -1 * min_stat_sum):

            # 가망 추가
            # 최대로 더할 수 있는 값이 -min_stat_sum보다 작음 -> 가망 X
            # 이미 값이 +min_stat_sum보다 큼 -> 가망 x
            tree.append([stat_sum, rest_team, step+1])

        # 마지막 학생 추가라면, 점수를 계산
        if(rest_team==1):
            score = abs(stat_sum + students[step])
            # print(node, score)

            if(score<min_stat_sum):
                min_stat_sum = score
            if(score<1):
                break

        #최소로 더할 수 있는 값이 min_stat_sum보다 큼 -> 가망 X
        elif(stat_sum + students[step] + sum(students[max_step-rest_team+1:max_step]) < min_stat_sum):
            tree.append([stat_sum + students[step], rest_team-1, step+1])

    return int(pivot*team_num + min_stat_sum) * int(pivot*team_num - min_stat_sum)
    

n, team_num, stat_sum = map(int, input().split())

pivot = stat_sum / 2
students = []
for _ in range(n):
    students.append(int(input().split()[0]) - pivot)

print(calc_stat(students, team_num, stat_sum))



```

```python
def calc_stat(students, team_num, stat_sum):
    
    mid = team_num * stat_sum // 2

    sums = [[] for _ in range(team_num)]

    for stat in students:

        for s in range(0, team_num-1):

            for n in sums[s+1]:
                sum_stat = n + stat
                if(sum_stat not in sums[s]):
                    sums[s].append(sum_stat)

        if(stat not in sums[team_num-1]):
            sums[team_num-1].append(stat)

    min = abs(sums[0][0] - mid)
    for stat_sums in sums[0]:
        tmp = abs(stat_sums - mid)
        if(tmp<min):
            min = tmp

    return (min+mid) * (team_num * stat_sum - (min+mid))


n, team_num, stat_sum = map(int, input().split())

students = []
for _ in range(n):
    students.append(int(input().split()[0]))

print(calc_stat(students, team_num, stat_sum))
```
