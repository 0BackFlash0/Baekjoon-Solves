```python
import sys
input = sys.stdin.readline

def riddle(riddle_map, n, m):
    # row, col, 벽을 부숨 여부
    pos_list = [(0, 0, 1)]
    riddle_map[0][0] = -2
    next_pos = []

    step = 1
    is_moveable = False

    while(pos_list):

        for pos in pos_list:
            r = pos[0]
            c = pos[1]
            breakable = pos[2]

            if(r==n-1 and c==m-1):
                is_moveable = True
                break

            #위 이동
            if(r!=0):
                if(-breakable<=riddle_map[r-1][c]<=0):
                    riddle_map[r-1][c] = -breakable-1
                    next_pos.append((r-1, c, breakable))
                elif(riddle_map[r-1][c]==1 and breakable>0):
                    next_pos.append((r-1, c, breakable-1))
            
            #아래 이동
            if(r!=n-1):
                if(-breakable<=riddle_map[r+1][c]<=0):
                    riddle_map[r+1][c] = -breakable-1
                    next_pos.append((r+1, c, breakable))
                elif(riddle_map[r+1][c]==1 and breakable>0):
                    next_pos.append((r+1, c, breakable-1))

            #왼쪽 이동
            if(c!=0):
                if(-breakable<=riddle_map[r][c-1]<=0):
                    riddle_map[r][c-1] = -breakable-1
                    next_pos.append((r, c-1, breakable))
                elif(riddle_map[r][c-1]==1 and breakable>0):
                    next_pos.append((r, c-1, breakable-1))
            
            #오른쪽 이동
            if(c!=m-1):
                if(-breakable<=riddle_map[r][c+1]<=0):
                    riddle_map[r][c+1] = -breakable-1
                    next_pos.append((r, c+1, breakable))
                elif(riddle_map[r][c+1]==1 and breakable>0):
                    next_pos.append((r, c+1, breakable-1))


        if(is_moveable):
            return step

        step += 1
        pos_list = next_pos

        # print("---------------------")
        # for col in riddle_map:
        #     print("  ".join(map(lambda i : "{:3d}".format(i), col)))
        # print("---------------------")
        # print(pos_list)
        del next_pos
        next_pos = []


    return -1




n, m = map(int, input().split())
riddle_map = []
for _ in range(n):
    riddle_map.append(list(map(int, input().replace("\n", ""))))

print(riddle(riddle_map, n, m))
```
