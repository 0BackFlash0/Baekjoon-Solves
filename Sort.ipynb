```python
import time, random
sample_list = random.sample(range(1, 15001), 15000)

# 정렬 여부 판정

def isSorted(check_list):
    sorted_list = sorted(check_list.copy())
    if(sorted_list==check_list):
        return True
    print(check_list)
    return False
```

```python
#1. 선택 정렬 (Selection Sort)

def selectionSort(sort_list):
    length = len(sort_list)
    for i in range(length):
        least_index = i
        for j in range(i+1, length):
            if(sort_list[least_index]>sort_list[j]):
                least_index = j
        sort_list[i], sort_list[least_index] = sort_list[least_index], sort_list[i]
```

```python
#2. 삽입 정렬 (Insertion Sort)

def insertionSort(sort_list):
    length = len(sort_list)
    for i in range(1, length):
        key = sort_list[i]
        j = i-1
        while (j>=0 and key<sort_list[j]):
            sort_list[j], sort_list[j+1] = sort_list[j+1], sort_list[j]
            j -= 1;
```

```python
#3. 버블 정렬 (Bubble Sort)

def bubbleSort(sort_list):
    length = len(sort_list)
    for i in range(0, length):
        for j in range(1, length-i):
            if(sort_list[j-1]>sort_list[j]):
                sort_list[j], sort_list[j-1] = sort_list[j-1], sort_list[j]
```

```python
#4 합병 정렬 (Merge Sort)

def mergeSort(sort_list):
    
    def divide(start, end):
        if(end-start<2):
            return
        
        mid = (start+end)//2

        divide(start, mid)
        divide(mid, end)

        merge(start, mid, end)
    
    
    def merge (start, mid, end):
        temp_list = []
        i = start
        j = mid

        while(i<mid and j<end):
            if(sort_list[i]<sort_list[j]):
                temp_list.append(sort_list[i])
                i += 1
            else:
                temp_list.append(sort_list[j])
                j += 1

        while (i<mid):
            temp_list.append(sort_list[i])
            i += 1
        while (j<end):
            temp_list.append(sort_list[j])
            j += 1

        for k in range(0, len(temp_list)):
            sort_list[start+k] = temp_list[k]
            
            
    return divide(0, len(sort_list))
```

```python
#퀵 정렬 (Quick Sort) - pivot을 첫 항목으로

def quickSort(sort_list):
    
    def sort(left, right):
        if(left >= right):
            return
        
        l, r = left+1, right
        pivot = sort_list[left]
        
        while(l<r):
            while(sort_list[r]>pivot and l<r):
                r -= 1

            while(sort_list[l]<pivot and l<r):
                l += 1
                
            if(l>=r):
                break
            
            sort_list[l], sort_list[r] = sort_list[r], sort_list[l]
            #print(pivot,"기준", sort_list[r], "<->", sort_list[l], sort_list[left:right+1])
        if(sort_list[l]<pivot):
            sort_list[l], sort_list[left] = sort_list[left], sort_list[l]
            #print("피벗", pivot, "위치 변경", sort_list[left:right+1])

        
        sort(left, l-1)
        sort(l, right)
        
        
    return sort(0, len(sort_list)-1)
```

#퀵 정렬 (Quick Sort) - in-place 정렬

def quickSort1(arr):
    def sort(low, high):
        if high <= low:
            return

        mid = partition(low, high)
        
        sort(low, mid-1)
        sort(mid, high)

    def partition(low, high):
        
        l, h = low, high
        pivot = arr[(low + high) // 2]
        
        while l <= h:
            while arr[l] < pivot:
                l += 1
            while arr[h] > pivot:
                h -= 1
            if l <= h:
                arr[l], arr[h] = arr[h], arr[l]
                #print(pivot,"기준", arr[h], "<->", arr[l], arr[low:high+1])
                l, h = l + 1, h - 1
            
            
        return l

    return sort(0, len(arr) - 1)



def quickSort2(arr):
    def sort(low, high):
        if high <= low:
            return

        mid = partition(low, high)
        
        sort(low, mid)
        sort(mid+1, high)

    def partition(low, high):
        
        l, h = low, high
        pivot = arr[(low + high) // 2 + (1 if (low+high)%2 else 0)]
        
        while l <= h:
            while arr[l] < pivot:
                l += 1
            while arr[h] > pivot:
                h -= 1
            if l <= h:
                arr[l], arr[h] = arr[h], arr[l]
                #print(pivot,"기준", arr[h], "<->", arr[l], arr[low:high+1])
                l, h = l + 1, h - 1
            
            
        return h

    return sort(0, len(arr) - 1)
